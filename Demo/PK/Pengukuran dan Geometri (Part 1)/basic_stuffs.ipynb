# Geometri Analitik dengan Matplotlib

Ini adalah contoh implementasi Geometri Analitik menggunakan Python dan Matplotlib. Kode ini menyelesaikan masalah geometri analitik terkait dengan dua titik dan persamaan garis.

## Fungsi-fungsi

### 1. Panjang Garis

Panjang garis antara dua titik $A(x_1, y_1)$ dan $B(x_2, y_2)$ dihitung menggunakan rumus jarak Euclidean:

$$ AB = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2} $$

### 2. Titik Tengah

Titik tengah ($M$) dari segmen garis $AB$ dihitung dengan menemukan rata-rata koordinat $x$ dan $y$ dari titik $A$ dan $B$:

$$ M\left(\frac{x_1 + x_2}{2}, \frac{y_1 + y_2}{2}\right) $$

### 3. Persamaan Garis

Persamaan garis yang melalui dua titik $A(x_1, y_1)$ dan $C(x_3, y_3)$ dihitung menggunakan rumus gradien ($m$) dan persamaan umum garis $y = mx + c$:

$$ m = \frac{y_3 - y_1}{x_3 - x_1} $$

$$ y = \left(\frac{y_3 - y_1}{x_3 - x_1}\right) \cdot x + \left(y_1 - \left(\frac{y_3 - y_1}{x_3 - x_1}\right) \cdot x_1\right) $$

```python
# Fungsi untuk menghitung panjang garis antara dua titik
def panjang_garis(x1, y1, x2, y2):
    return np.sqrt((x2 - x1)**2 + (y2 - y1)**2)

# Fungsi untuk menghitung titik tengah segmen garis
def titik_tengah(x1, y1, x2, y2):
    return ((x1 + x2) / 2, (y1 + y2) / 2)

# Fungsi untuk menghitung persamaan garis yang melalui tiga titik
def persamaan_garis(x1, y1, x3, y3):
    m = (y3 - y1) / (x3 - x1)
    c = y1 - m * x1
    return lambda x: m * x + c
```

```python
# titik-titik yang akan digunakan
A = (0, 0)
B = (5, 10)
C = (8, 2)
```

```python
# Plotting
plt.figure(figsize=(8, 6))

# Plot titik A, B, dan C
plt.scatter(*A, color='red', label='Titik A')
plt.scatter(*B, color='green', label='Titik B')
plt.scatter(*C, color='blue', label='Titik C')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.title('Plot Titik A, B, dan C')
plt.show()
```

```python
# Hitung panjang garis AB
panjang_AB = panjang_garis(*A, *B)

# Hitung titik tengah segmen garis AB
M_x, M_y = titik_tengah(*A, *B)

# Persamaan garis yang melalui dua titik
f = persamaan_garis(*A, *C)
```

```python
# Plotting
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# Plot in axs[0]
axs[0].scatter(*A, color='red', label='Titik A')
axs[0].scatter(*B, color='green', label='Titik B')
axs[0].scatter(*C, color='blue', label='Titik C')

axs[0].set_xlabel('X-axis')
axs[0].set_ylabel('Y-axis')
axs[0].axhline(0, color='black', linewidth=0.5)
axs[0].axvline(0, color='black', linewidth=0.5)
axs[0].grid(color='gray', linestyle='--', linewidth=0.5)
axs[0].legend()
axs[0].set_title('Plot Titik A, B, dan C')

# Plot in axs[1]
x = np.linspace(-1, 10, 100)
y_AC = f(x)

axs[1].plot([A[0], B[0]], [A[1], B[1]], label='Garis AB', color='blue', linestyle='--')
axs[1].scatter(*A, color='red', label='Titik A')
axs[1].scatter(*B, color='green', label='Titik B')
axs[1].scatter(*C, color='purple', label='Titik C')
axs[1].scatter(M_x, M_y, color='orange', label='Titik Tengah M')
axs[1].plot(x, y_AC, label='Garis AC', linestyle='--', color='magenta')

axs[1].set_xlabel('X-axis')
axs[1].set_ylabel('Y-axis')
axs[1].axhline(0, color='black', linewidth=0.5)
axs[1].axvline(0, color='black', linewidth=0.5)
axs[1].grid(color='gray', linestyle='--', linewidth=0.5)
axs[1].set_title('Plot Garis AB dan AC')
axs[1].legend()

plt.show()
```

* Rumus Persamaan Garis dengan 3 titik ada?

```python
# Extracting coordinates
x_values = [A[0], B[0], C[0]]
y_values = [A[1], B[1], C[1]]

# Solving for the coefficients of the quadratic equation (y = ax^2 + bx + c)
coefficients_quad = np.polyfit(x_values, y_values, 2)
a_quad, b_quad, c_quad = coefficients_quad

# Solving for the coefficients of the cubic equation (y = ax^3 + bx^2 + cx + d)
coefficients_cubic = np.polyfit(x_values, y_values, 3)
a_cubic, b_cubic, c_cubic, d_cubic = coefficients_cubic

# Solving for the coefficients of the quintic equation (y = ax^5 + bx^4 + cx^3 + dx^2 + ex + f)
coefficients_quintic = np.polyfit(x_values, y_values, 5)
a_quintic, b_quintic, c_quintic, d_quintic, e_quintic, f_quintic = coefficients_quintic

# Generating x values for the plot
x = np.linspace(min(x_values), max(x_values), 100)

# Calculating y values using the quadratic, cubic, and quintic equations
y_quad = a_quad * x**2 + b_quad * x + c_quad
y_cubic = a_cubic * x**3 + b_cubic * x**2 + c_cubic * x + d_cubic
y_quintic = a_quintic * x**5 + b_quintic * x**4 + c_quintic * x**3 + d_quintic * x**2 + e_quintic * x + f_quintic

# Creating a subplot with 1 row and 1 column
fig, axs = plt.subplots(1, 1, figsize=(15, 6))

# Plotting the points and the parabolic, cubic, and quintic trajectories
axs.plot(x, y_quad, label=f'Parabolic: y = {a_quad:.10f}x^2 + {b_quad:.10f}x + {c_quad:.10f}')
axs.plot(x, y_cubic, label=f'Cubic: y = {a_cubic:.10f}x^3 + {b_cubic:.10f}x^2 + {c_cubic:.10f}x + {d_cubic:.10f}')
axs.plot(x, y_quintic, label=f'Quintic: y = {a_quintic:.10f}x^5 + {b_quintic:.10f}x^4 + {c_quintic:.10f}x^3 + {d_quintic:.10f}x^2 + {e_quintic:.10f}x + {f_quintic:.10f}')

# Adding labels and legend
axs.scatter(x_values, y_values, label='Given Points', color='red')
axs.set_title('Parabolic, Cubic, and Quintic Trajectories through Given Points')
axs.set_xlabel('x-axis')
axs.set_ylabel('y-axis')
axs.legend()
axs.grid(color='gray', linestyle='--', linewidth=0.5)

# Display the plot
plt.show()
```
